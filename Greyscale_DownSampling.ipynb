```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### ***Apply grayscale resolution on the test image to obtain new image***

s= (2b-1/2k-1)âˆ—r

where, 
* b=old bits
* k=new bits

```python
img1 = cv2.imread('/content/cameraman.tif',0) #reading the image
type(img1)
```

```python
np.max(img1) #finding the maximum pixel value
```

```python
[m,n]= img1.shape #stores the shape of the image
```

```python
k = int(input("Enter no. of required bits for the new image = "))
```

```python
original = 256 #2^8
new = (2**k)-1
Constant = new/(original-1)
Constant
```

```python
#Changing bits
img1_gr= np.zeros((m,n),dtype= np.int)

for i in range(m):
  for j in range(n):

    img1_gr[i][j] = Constant*img1[i][j]
```

```python
img1_gr
```

```python
#Plotting and comparing both the images
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(img1,cmap ='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img1_gr,cmap ='gray')
plt.title('Reduced Grayscale Resolution')
```

```python
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(img1)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img1_gr)
plt.title('Reduced Resolution')
```

### ***Apply sampling resolution on the test image to obtain the new image***

```python
print("Shape of original image:", img1.shape)
```

```python
rate = int(input("enter the down sampling rate: ")) #factor to downsize
```

```python
img1_ds = img1[::rate,::rate]  #slicing the image 
```

```python
print("Shape of downsized image:", img1_ds.shape)
```

```python
#Plotting and comparing the original and the downsized image
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.imshow(img1, cmap ='gray')
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(img1_ds, cmap ='gray')
plt.title('Downsized')
```
