```python

import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
#Method 1 Generate unit ramp signal
n = np.arange(-10,11)
r = np.where(n>=0,n,0) #r=n for n>0
print(r)
print(n)
plt.stem(n,r)
```

```python
#Method 2
n1 = np.linspace(-5,10,16)
r1 =n1*(n1>=0)
print(n1)
print(r1)
plt.stem(n1,r1)
```

```python
#unit step
n2 = np.arange(-10,11)
r2 = np.where(n>=0,1,0)
print(r2)
print(n2)
plt.stem(n2,r2)
```

```python
#impulse
l = [1 for n in range(-20,10) if n==1]
print(l)
plt.stem(l)
```

```python
#exponential 
plt.figure(figsize=(14,14))
n = np.arange(0,11,1)
a = 0.5
x = [a**i for i in n]
print(x)
plt.subplot(2,2,1)
plt.stem(n,x)
plt.title('For 0<a<=1')

a = 1.5
x = [a**i for i in n]
print(x)
plt.subplot(2,2,2)
plt.stem(n,x)
plt.title('For a>1')

n = np.arange(0,11,1)
a = -0.5
x = [a**i for i in n]
print(x)
plt.subplot(2,2,3)
plt.stem(n,x)
plt.title('For -1<a<0')

n = np.arange(0,11,1)
a = -1.5
x = [a**i for i in n]
print(x)
plt.subplot(2,2,4)
plt.stem(n,x)
plt.title('For a<-1')
```

```python
#Time Shifting
n = np.arange(-4,6)
x = [0,0,0,0,2,1,2,1,0,0]
k = 2 
fig,axis = plt.subplots(3,1, figsize=(10,12))
axis[0].stem(n,x)
axis[1].stem(n+k,x) 
axis[2].stem(n-k,x) 
```

```python
#Time Reversal
n = np.arange(-4,6)
x = [0,0,0,0,2,1,2,1,0,0]
fig,axis = plt.subplots(1,2, figsize=(14,4))
axis[0].stem(n,x)
axis[1].stem(-1*n,x)
```

```python
#Amplitude Scaling
plt.figure(figsize=(21,5))

# y(n) = x(n)
plt.subplot(1,3,1)
n = np.arange(-4,6)
x = [0,0,0,0,2,1,2,1,0,0]
plt.stem(n,x)
plt.title('Original')
plt.xlabel('n')
plt.ylabel('x(n)')

# y(n) = 2x(n)
plt.subplot(1,3,2)
n = np.arange(-4,6)
x = [0,0,0,0,2,1,2,1,0,0]
x = [2*i for i in x]
plt.stem(n,x)
plt.title('Amplifying')
plt.xlabel('n')
plt.ylabel('x(n)')

# y(n) = x(n)/2
plt.subplot(1,3,3)
n = np.arange(-4,6)
x = [0,0,0,0,2,1,2,1,0,0]
x = [i/2 for i in x]
plt.stem(n,x)
plt.title('Attenuation')
plt.xlabel('n')
plt.ylabel('x(n)')
```

```python
#Time Scaling
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
# y(n) = x(2n)
a = [1,2,3,4,5,4,3,2,1]
n = np.arange(0,9,1)
new_n = [2*x for x in n]
y = [a[x] if x in new_n else 0 for x in n]
plt.stem(n,y)
plt.title('Downsampling')

plt.subplot(1,2,2)
# y(n) = x(n/2)
a = [1,2,3,4,5,4,3,2,1]
n = np.arange(0,18,1)
new_n = [int(x/2) if x%2==0 else '-' for x in n]
y = [a[x] if type(x)==int else 0 for x in new_n]
plt.stem(n,y)
plt.title('Upsampling or Interpolation')
```

```python
#Discrete, Continuous and Digital signals
plt.figure(figsize=(21,5))

# Continous timed and Continuous Valued
plt.subplot(1,3,1)
t = np.arange(0,1,0.01)
f = 2
A = 1
x = A * np.sin(2*np.pi*f*t)
plt.plot(t,x)
plt.title('Continous timed and Continuous Valued')
plt.xlabel('t')
plt.ylabel('x')

# Discrete timed and Continuous Valued
plt.subplot(1,3,2)
t = np.arange(0,1,0.1)
f = 2
A = 1
x = A * np.sin(2*np.pi*f*t)
plt.stem(t,x)
plt.title('Discrete timed and Continuous Valued')
plt.xlabel('t')
plt.ylabel('x')

# Discrete timed and Discrete Valued
plt.subplot(1,3,3)
t = np.arange(0,1,0.1)
f = 2
A = 1
x = A * np.sin(2*np.pi*f*t)
x = [np.round(i) for i in x]
plt.stem(t,x)
plt.title('Discrete timed and Discrete Valued')
plt.xlabel('t')
plt.ylabel('x')
```
